```python
import numpy as np
```

```python
x = np.linspace(0,2,10)
y = x**2
print(x)
```

**1.Pyplot**

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,c='red',lw=3,ls='--')
plt.show()
```

```python
plt.plot(x,y,c='green',lw=2,ls=':')
plt.show()
```

```python
plt.figure()
plt.plot(x,y,label='quadratique')
plt.plot(x,x**3,label='cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure.png')
```
