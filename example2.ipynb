```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2,10)
y = x**2
print(x)
```

```python
#OOP method
fig,ax = plt.subplots()
ax.plot(x,y)
plt.show()
```
